```python
df = pd.read_csv("rlog_transformed_data.csv", sep=",")
df = df.dropna()
df.head()
```

```python
df.rename(columns={'Unnamed: 0': 'gene'}, inplace=True)
df.head()
```

```python
X = df.set_index('gene').T
y = X.index
X.head()
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
X_reduced = pca.fit_transform(X)
```

```python
import matplotlib.pyplot as plt

plt.scatter(X_reduced[:, 0], X_reduced[:, 1], cmap='viridis')
plt.xlabel('Ð¡omponent 1')
plt.ylabel('Ð¡omponent 2')
#plt.colorbar(label='ÐšÐ»Ð°ÑÑ‚ÐµÑ€Ñ‹')
plt.show()
```

```python
import plotly.express as px

pca_df = pd.DataFrame(data=X_reduced, columns=['PC1', 'PC2'])
pca_df['label'] = y
pca_df['group'] = pca_df['label'].apply(lambda x: '_'.join(x.split("_")[:-1]))
print(pca_df.head())

fig = px.scatter(
    pca_df,
    x='PC1',
    y='PC2',
    hover_data=['label'],
    color='group',
)

fig.update_layout(
    plot_bgcolor='white',
    xaxis=dict(
        linecolor='black',
        gridcolor='lightgrey'
    ),
    yaxis=dict(
        linecolor='black',
        gridcolor='lightgrey'
    )
)
fig.show()
```

```python
from sklearn.cluster import KMeans
```

```python
n=6
kmeans = KMeans(n_clusters=n, init='k-means++')
labels_kmeans = kmeans.fit_predict(X_reduced)
```

```python
plt.scatter(X_reduced[:, 0], X_reduced[:, 1], c=labels_kmeans, cmap='viridis')
plt.xlabel('Ð¡omponent 1')
plt.ylabel('Ð¡omponent 2')
plt.colorbar(label='Clusters')
plt.show()
```

```python
pca_df = pd.DataFrame(data=X_reduced, columns=['PC1', 'PC2'])
pca_df['label'] = y
pca_df['cluster'] = labels_kmeans
pca_df['group'] = pca_df['label'].apply(lambda x: '_'.join(x.split("_")[:-1]))
print(pca_df.head())

fig = px.scatter(
    pca_df,
    x='PC1',
    y='PC2',
    hover_data=['label'],
    color='group',
    symbol='cluster'
)

fig.update_layout(
    plot_bgcolor='white',
    xaxis=dict(
        linecolor='black',
        gridcolor='lightgrey'
    ),
    yaxis=dict(
        linecolor='black',
        gridcolor='lightgrey'
    )
)

fig.show()
```

```python
df_sign_genes = pd.read_csv("deseq2_results_significant_genes.csv", sep=",")
df_sign_genes = df_sign_genes.dropna()
df_sign_genes.head()
```

```python
sign_genes = df_sign_genes['Unnamed: 0'].tolist()
filtered_df = df[df['gene'].isin(sign_genes)]
print(len(df))
print(len(filtered_df))
filtered_df.head()
```

```python
X = filtered_df.set_index('gene').T
y = X.index

pca = PCA(n_components=2)
X_reduced = pca.fit_transform(X)


pca_df = pd.DataFrame(data=X_reduced, columns=['PC1', 'PC2'])
pca_df['label'] = y
pca_df['group'] = pca_df['label'].apply(lambda x: '_'.join(x.split("_")[:-1]))
print(pca_df.head())

fig = px.scatter(
    pca_df,
    x='PC1',
    y='PC2',
    hover_data=['label'],
    color='group'
)

fig.update_layout(
    plot_bgcolor='white',
    xaxis=dict(
        linecolor='black',
        gridcolor='lightgrey'
    ),
    yaxis=dict(
        linecolor='black',
        gridcolor='lightgrey'
    )
)
fig.show()
```

```python
n=5
kmeans = KMeans(n_clusters=n, init='k-means++')
labels_kmeans = kmeans.fit_predict(X_reduced)

pca_df = pd.DataFrame(data=X_reduced, columns=['PC1', 'PC2'])
pca_df['label'] = y
pca_df['cluster'] = labels_kmeans
pca_df['group'] = pca_df['label'].apply(lambda x: '_'.join(x.split("_")[:-1]))
print(pca_df.head())

fig = px.scatter(
    pca_df,
    x='PC1',
    y='PC2',
    hover_data=['label'],
    color='group',
    symbol='cluster'
)
fig.update_layout(
    plot_bgcolor='white',
    xaxis=dict(
        linecolor='black',
        gridcolor='lightgrey'
    ),
    yaxis=dict(
        linecolor='black',
        gridcolor='lightgrey'
    )
)
fig.show()
```

```python
from sklearn.cluster import DBSCAN

dbscan = DBSCAN(eps=0.1, min_samples=2)
labels_dbscan = dbscan.fit_predict(X_reduced)

print(max(labels_dbscan))

pca_df = pd.DataFrame(data=X_reduced, columns=['PC1', 'PC2'])
pca_df['label'] = y
pca_df['cluster'] = labels_dbscan
print(pca_df.head())

fig = px.scatter(
    pca_df,
    x='PC1',
    y='PC2',
    hover_data=['label'],
    color='cluster'
)
fig.update_layout(
    plot_bgcolor='white',
    xaxis=dict(
        linecolor='black',
        gridcolor='lightgrey'
    ),
    yaxis=dict(
        linecolor='black',
        gridcolor='lightgrey'
    ),
    showlegend=False
)
fig.show()
```

```python
dbscan = DBSCAN(eps=0.1, min_samples=2)
labels_dbscan = dbscan.fit_predict(X)

print(max(labels_dbscan))
```

```python
from sklearn.cluster import HDBSCAN

hdb = HDBSCAN(min_cluster_size=2, min_samples=1)
hdb_labels = hdb.fit_predict(X_reduced)

print(max(hdb_labels))
print(list(hdb_labels).count(-1))

pca_df = pd.DataFrame(data=X_reduced, columns=['PC1', 'PC2'])
pca_df['label'] = y
pca_df['cluster'] = hdb_labels
print(pca_df.head())

fig = px.scatter(
    pca_df,
    x='PC1',
    y='PC2',
    hover_data=['label'],
    color='cluster'
)
fig.update_layout(
    plot_bgcolor='white',
    xaxis=dict(
        linecolor='black',
        gridcolor='lightgrey'
    ),
    yaxis=dict(
        linecolor='black',
        gridcolor='lightgrey'
    ),
    showlegend=False
)
fig.show()
```
